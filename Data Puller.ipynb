```python
#Import dependencies
import os
import requests
import json
import pandas as pd
from pprint import pprint
from config import api_key
```

```python

##URLs for all API query's

# Base URL for Covid Data
query_tx_url="https://api.covidtracking.com/v1/states/tx/daily.csv"
query_ny_url="https://api.covidtracking.com/v1/states/ny/daily.csv"
query_all_url="https://api.covidtracking.com/v1/states/daily.csv"


# BLS base URL for Houston, Teen, Men, Women Unemployment Data
bls_url = "https://api.bls.gov/publicAPI/v2/timeseries/data/"

# Header BLS uses for API's
headers = {"Content-type":"application/json"}
```

```python

# Downloaded Texas Labor Market Data Files
fileTLM1_to_load = "Resources/Texas Labor Market CES Data.csv"
fileTLM2_to_load = "Resources/Texas Labor Market Indicators.csv"
fileTLM3_to_load = "Resources/Texas Labor Market LAUS Data.csv"
```

```python

# BLS API query for Houston unemployment

headers = {"Content-type":"application/json"}
data = json.dumps({"seriesid":["LAUMT482642000000004"],"startyear":"2019", "endyear":"2020","registrationkey":api_key})
Houston_unempl = requests.post(bls_url, data=data, headers=headers)

# Pretty Print the output of the JSON
response = (Houston_unempl).json()

length = response["Results"]["series"][0]["data"]

#Drill down into data
unemployment=[]
month = []
year=[]

#
for rows in length:
    unemployment.append(rows["value"]) 
    month.append(rows["periodName"]) 
    year.append(rows["year"]) 



Houston_df = pd.DataFrame({"Month": month,
                           "Year": year,
                           "Tot Claims": unemployment})

Houston_df=Houston_df.sort_index(ascending=False)

Houston_df.head()
```

```python

#Export dataframe to csv output file
output_file = os.path.join("Resources","Houston_data.csv")
Houston_df.to_csv(output_file, index=False, header=True)
```

```python

# BLS API query for Teens unemployment

headers = {"Content-type":"application/json"}
data = json.dumps({"seriesid":["LNS14000012"],"startyear":"2019", "endyear":"2020","registrationkey":api_key})
Teen_unempl = requests.post(bls_url, data=data, headers=headers)

# Pretty Print the output of the JSON
response = (Teen_unempl).json()
length = response["Results"]["series"][0]["data"]
# pprint(response)
#Drill down into data
Teen_unemployment=[]
month = []
year=[]

#
for rows in length:
    Teen_unemployment.append(rows["value"]) #= response["Results"]["series"][0]["data"][0]["value"]
    month.append(rows["periodName"]) #= response["Results"]["series"][0]["data"][0]["periodName"]
    year.append(rows["year"]) #= response["Results"]["series"][0]["data"][0]["year"]


Teen_df = pd.DataFrame({ "Month": month,
                        "Year": year,
                        "% Teen Unempl": Teen_unemployment})

Teen_df=Teen_df.sort_index(ascending=False)

Teen_df.head()
```

```python

#Export dataframe to csv output file
output_file = os.path.join("Resources","Teen_unemplyed.csv")
Teen_df.to_csv(output_file, index=False, header=True)
```
